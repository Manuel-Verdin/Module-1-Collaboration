```python
secret = 7
guess = 5

if guess < secret:
    print('too low')
elif guess > secret:
    print('too high')
else:
    print('just right')
```

4.2

```python
small = True
green = False

if small and green:
    print('pea')
elif small and not green:
    print('cherry')
elif not small and green:
    print('watermelon')
else:
    print('pumpkin')
```

6.2

```python
guess_me = 7
number = 1

while True:
    if number < guess_me:
        print('too low')
    elif number == guess_me:
        print('found it!')
        break
    else:
        print('oops')
        break

    number += 1
```

```python
guess_me = 5

for number in range(10):
    if number < guess_me:
        print('too low')
    elif number == guess_me:
        print('found it!')
        break
    else:
        print('oops')
        break
```
